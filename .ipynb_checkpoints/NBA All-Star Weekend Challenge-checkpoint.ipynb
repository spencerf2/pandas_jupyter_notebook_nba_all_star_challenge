# What is the average weight of all players who played during this time?

```python
import numpy as np
import pandas as pd

nba = pd.read_csv('NBA All Stars 2000-2016 - Sheet1.csv')
display(nba)

display(nba[['WT']].mean().round(decimals = 2).to_frame())
```

# What team is represented the most during this time?

```python
nba_max_represented = nba.copy()

nba_max_represented['Most Represented Team'] = 1

display(nba_max_represented.groupby(['Team'])[['Most Represented Team']].count().sort_values(by='Most Represented Team', ascending = False).head(1))

display(nba_max_represented)
```

# What draft class is represeted the most during this time?

```python
nba_max_represented_draft_class = nba.copy()

nba_max_represented_draft_class['Most Represented Draft Class'] = 1

display(nba_max_represented_draft_class.groupby(['NBA Draft Status'])[['Most Represented Draft Class']].count().sort_values(by='Most Represented Draft Class', ascending = False).head(1))

display(nba_max_represented_draft_class)
```

# Are foriegn players more prevalent in the All-Star Game during this time?

```python
us_players = nba[nba['Nationality'] == 'United States']
foreign_players = nba[nba['Nationality'] != 'United States']

total_us_players = us_players['Nationality'].count()
total_foreign_players = foreign_players['Nationality'].count()

if total_foreign_players > total_us_players:
    print(f'Yes, there were {total_foreign_players - total_us_players} more foreign players.')
elif total_us_players == total_foreign_players:
    print(f'There is an equal number of players from inside and outside The United States.')
else:
    print(f'No, The US had {total_us_players - total_foreign_players} more players during this time.')
```

# How often are the Western All-Stars voted in by fans?

```python
west_fan_vote = nba[nba['Selection Type'] == 'Western All-Star Fan Vote Selection']

print(f"They were voted in {west_fan_vote['Nationality'].count()} times over this period.")
```

# How many times does Steph Curry make the All-Star game during this time period?

```python
steph_curry = nba[nba['Player'] == 'Stephen Curry']

print(f"Stephen Curry made it {steph_curry['Player'].count()} times.")
```

# How many Shooting Guards have made the All-Star game during this time?

```python
shooting_guards = nba[nba['Pos'] == 'SG']

print(f"There were {shooting_guards['Pos'].count()} shooting guards during this time.")
```

# How many Chicago Bulls players have made the All-Star Game during this time period?

```python
chicago_bulls = nba[nba['Team'] == 'Chicago Bulls']

print(f"{chicago_bulls['Team'].count()} Chicago Bulls players made the All-Star Game during this time.")
```
